```python
# handle missing values
import numpy as np
import pandas as pd
data = {
    "experience":[1,np.nan,3,4,5],
    "salary":[100,200,300,400,np.nan],
    "name":['hello','dheeraj','Nikhil','Raj','vinay']
}
df = pd.DataFrame(data)
print(df)
# null value filter | find out
print(df.isnull())

# sum of null
print(df.isnull().sum())

# dropna | drop complete row
df.dropna()
print(df)

# fillna | null -> fill
df.fillna(5)
```

```python
# pie chart
import matplotlib.pyplot as plt #

x = ["apple","banana","carrot","mango"]
y = [30,50,70,5]

plt.figure(figsize=(4,4))
plt.pie(y,labels=x,autopct="%1.1f%%",colors=["red","yellow","orange","green"],shadow=True)
plt.show()
```

```python
# sub plot
import matplotlib.pyplot as plt #

# pie chart data
x = ["apple","banana","carrot","mango"]
y = [30,50,70,5]
plt.subplot(1,2,1)
plt.pie(y,labels=x)

# line chart data
data = {
    "experience":[1,2,3,4,5],
    "salary":[100,200,300,400,500]
}
x = data['experience']
y = data['salary']
plt.subplot(1,2,2)
plt.plot(x,y)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ["Apple", "Banana", "Orange", "Mango", "Kiwi",
     "Grapes", "Papaya", "Guava", "Pear", "Litchi"]

y = [18, 25, 32, 40, 22, 35, 28, 30, 20, 27]

plt.figure(figsize=(12, 8))

plt.subplot(2,2,1)
plt.pie(y, labels=x)

plt.subplot(2,2,2)
plt.bar(x, y)
plt.xticks(rotation=45)

plt.subplot(2,2,3)
plt.plot(x, y, marker='o')
plt.xticks(rotation=45)

plt.subplot(2,2,4)
plt.barh(x, y)

plt.tight_layout()
plt.show()
```
